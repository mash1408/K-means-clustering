```python
from sklearn.datasets import load_iris
import pandas as pd
import math
import matplotlib.pyplot as plt

clusters=[]
iris=load_iris()
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
def euclideanDistance(p1,p2):
  sum=0
  for key in p1:
    sum+=pow(p1[key]-p2[key],2)
  return math.sqrt(sum)

def Centroid(p):
   centroid={}
   centroid['sepal length (cm)']=p[-2]['sepal length (cm)']/4
   centroid['sepal width (cm)']=p[-2]['sepal width (cm)']/4
   centroid['petal length (cm)']=p[-2]['petal length (cm)']/4
   centroid['petal width (cm)']=p[-2]['petal width (cm)']/4
   return centroid

def Kmeans(k):
  for i in range(k):
    clusters.insert(i,[])
    # Initialise the k clusters
  for i in range(k):
    clusters[i].append(iris_df.iloc[i].to_dict())
    clusters[i].append({})
    clusters[i][-1]=Centroid(clusters[i])
#Iterate through the iris_df
  for i in range(3,25):
    select=0
    dist=[]
  # find euclidean distance for cluster's centroid and point  
    for cluster in clusters:
      dist.append(euclideanDistance(cluster[-1],iris_df.iloc[i].to_dict()))
    min_index = dist.index(min(dist))
    clusters[min_index].insert(-2,iris_df.iloc[i].to_dict())
    clusters[min_index][-1]=Centroid(clusters[min_index])
Kmeans(3)

c1 = pd.DataFrame(clusters[0])
c2 = pd.DataFrame(clusters[1])
c3 = pd.DataFrame(clusters[2])

c1.plot(kind ="scatter",
          x ='sepal length (cm)',
          y ='petal length (cm)',c = 'red', label = 'Cluster1')
c2.plot(kind ="scatter",
          x ='sepal length (cm)',
          y ='petal length (cm)',c = 'green', label = 'Cluster2')
c3.plot(kind ="scatter",
          x ='sepal length (cm)',
          y ='petal length (cm)',c = 'yellow', label = 'Cluster3')
plt.grid()
```
